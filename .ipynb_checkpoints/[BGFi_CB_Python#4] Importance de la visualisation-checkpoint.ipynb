```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
random_data = np.random.randn(50, 100)

random_data[abs(random_data) > 2] = np.nan

df_missing = pd.DataFrame(random_data)

df_missing.iloc[12, 60:] = np.nan
df_missing.iloc[23, 20:41] = np.nan
df_missing.iloc[23, 80:90] = np.nan
df_missing.iloc[15, 90:] = np.nan
df_missing.iloc[28, 71:] = np.nan


df_missing.iloc[:, 31] = np.nan
df_missing.iloc[:, 18] = np.nan
```

```python
plt.figure(figsize=(18, 9))
sns.heatmap(df_missing.isnull(), cbar=False)
plt.show()
```

# Valeurs extrêmes (outliers)

```python
data_random = np.random.randn(10000)
sns.distplot(data_random)
```

```python
plt.figure(figsize=(12, 4))
sns.boxplot(data=data_random, orient='h')
```

Box plots avec PlotLy

```python
import plotly.express as px
df = px.data.tips()
fig = px.box(df, x="time", y="total_bill")
fig.show()
```

```python
import plotly.express as px
df = px.data.tips()
fig = px.box(df, x="time", y="total_bill", points="all")
fig.show()
```

```python
df_quartiles = np.sort(df[df["time"] == "Dinner"]["total_bill"])
print(df_quartiles)
```

```python
q1 = int(len(df_quartiles) / 4)
q3 = 3*int(len(df_quartiles) / 4)
print(f"Premier quartile q1 : {q1}")
print(f"Troisième quartile q3 : {q3}")
```

```python
print("Valeurs comprises dans l'écart interquartile Q1-Q3:")
print(df_quartiles[q1:q3])
```
